```python
#1. is_prime(n) funksiyasini hosil qiling(n>0). Agar n soni tub bo'lsa True, aks holda False qiymatini qaytarsin.

def is_prime(n): 
    if n<=1:
        return False
    range_list = list(range(2, int(n**0.5) + 1))
    for i in range_list:
        if n%i==0:
            return False
    return True
    
    
    
print(is_prime(4))
print(is_prime(7))

```

```python
#2. k sonining raqanlari yig'indisini hisoblovchi, digit_sum(k) rekursiv funksiya tuzing

def digit_sum(k):
    raqamlar_yigindisi=0
    for i in str(k):
        raqamlar_yigindisi+=int(i)
    return raqamlar_yigindisi

digit_sum(15)
```

```python
#3. It is required to print all integer powers of two(that is, numbers of form 2**k) that do not exceed number N.

def func1(n):
    k=0
    while k**2<=n:
        print(f"{k} - {k**2}")
        k+=1
    
func1(121)
```
